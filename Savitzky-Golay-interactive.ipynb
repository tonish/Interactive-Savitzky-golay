```python
@interact(n=widgets.IntSlider(min=1, max=30, step=2, value=3),
          polyorder=widgets.IntSlider(min=1, max=3, step=1, value=1))



def SavitzkyGolay(n,polyorder,d=0):
    """n is the window width, polyorder is the polynom"""
    SG_df = SG(a,n, polyorder, deriv=d)
    plt.figure(figsize=(15,10))
    plt.plot(bands,a,linewidth=2)
    plt.plot(bands,SG_df,linewidth=2)
```

```python
@interact(soft=widgets.IntSlider(min=1, max=10, step=1, value=1),
         beta=widgets.IntSlider(min=1, max=30, step=1, value=10),
         width=widgets.IntSlider(min=1, max=30, step=1, value=9),
         )

def bil(soft,beta,width):
    # do bilateral filter
    plt.figure(figsize=(15,10))
    df_bl=bilateral(a,soft=soft, beta=beta, width=width, display=0) 
    plt.plot(bands,a,linewidth=3)
    plt.plot(bands,df_bl,linewidth=3)
```
